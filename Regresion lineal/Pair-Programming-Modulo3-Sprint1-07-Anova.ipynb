```python
df = pd.read_csv("Datos/possum_esta.csv", index_col = 0)
df.head()
```

Hipótesis nula (HO). No hay relación entre el tamaño de la cabeza y otras partes del cuerpo

Hipótesis alternativa (H1). Hay relación entre el tamaño de la cabeza y otras partes del cuerpo (cráneo, dientes, torso)

```python
lm = ols('hdlngth ~ skullw + totlngth + taill + footlgth + earconch + eye + chest + belly + age' , data=df).fit()
sm.stats.anova_lm(lm)
```

```python
lm.summary()
```

Skullw y totlngth son significativas, el pvlue es menor que 0.05, rechazamos la hipótesis nula
